```python
import re
import pandas as pd
```

```python
df = pd.read_csv('test_data.csv')
df
```

```python
begin = re.compile('[\[\{]+')
end = re.compile('[\}\]]+')
series = []
for n, sent in enumerate(df['sentence']):
    if begin.findall(sent):
        sent = begin.sub('(', sent)
        sent = end.sub(')', sent)
    series.append(sent)
```

```python
df['sentence'] = pd.Series(series)
df
```

```python
for sent in df['sentence']:
    if begin.findall(sent):
        print(sent)
```

```python
df.to_csv('modified_test.csv', index=False)
```
