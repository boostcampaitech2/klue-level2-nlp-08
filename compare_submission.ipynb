```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
output_dfs = []
for i in range(1, 11):
    output_dfs.append(pd.read_csv(f'../klue-level2-nlp-08/submissions/output ({i}).csv'))
```

```python
train_df = pd.read_csv('../klue-level2-nlp-08/data/train.csv')
```

```python
normalize = output_dfs[0].shape[0]/train_df.shape[0]
```

## without no relation

```python
label_list = ['no_relation','org:member_of','org:top_members/employees','org:alternate_names', 'per:date_of_birth', 'org:place_of_headquarters', 'per:employee_of', 'per:origin', 'per:title', 'org:members', 'per:schools_attended', 'per:colleagues', 'per:alternate_names', 'per:spouse', 'org:founded_by', 'org:political/religious_affiliation', 'per:children', 'org:founded', 'org:number_of_employees/members', 'per:place_of_birth', 'org:dissolved', 'per:parents', 'per:religion', 'per:date_of_death', 'per:place_of_residence', 'per:other_family', 'org:product', 'per:siblings', 'per:product', 'per:place_of_death']

fig = plt.figure(figsize=(20, 13))

palette = list(reversed(sns.color_palette("RdPu", len(output_dfs))))
for idx, output_df in enumerate(output_dfs):
    label_counts = output_df['pred_label'].value_counts()
    counts_dict = {label:label_counts.get(label, 0) for label in label_list}
    counts_df = pd.DataFrame.from_dict(counts_dict, orient='index', columns=['cnt']).drop('no_relation')
    g = sns.lineplot(data=counts_df, x=counts_df.index, y='cnt', color=palette[idx], label=idx)

    plt.xticks(rotation=45, ha='right')

    plt.gca().spines['right'].set_visible(False) 
    plt.gca().spines['top'].set_visible(False)

label_counts = train_df['label'].value_counts()*normalize
counts_dict = {label:label_counts.get(label, 0) for label in label_list}
counts_df = pd.DataFrame.from_dict(counts_dict, orient='index', columns=['cnt']).drop('no_relation')
g = sns.lineplot(data=counts_df, x=counts_df.index, y='cnt', color='black', label='train')


plt.grid()
plt.show()
```
